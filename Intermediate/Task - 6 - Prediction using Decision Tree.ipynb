```python
import pandas as pd 
import sklearn.datasets as df 
```

```python
iris = df.load_iris()
```

```python
x = pd.DataFrame(iris.data, columns = iris.feature_names)
x.head()
```

```python
x.info()
```

```python
x.describe()
```

```python
x.isnull().sum()
```

```python
y = iris.target
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42) 
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
a = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
b = ['setosa','versicolor','virginica']

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (5,5), dpi = 300)
tree.plot_tree(dt, feature_names = a, class_names = b, filled = True);
```
